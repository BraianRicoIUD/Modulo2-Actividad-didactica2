```python
# ============================================================
# ACTIVIDAD DIDÁCTICA 2 - MODELO OCULTO DE MARKOV
# Simulación del clima a partir de observaciones de humedad
# Código preparado para Google Colab
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Para que los resultados sean reproducibles
np.random.seed(42)

# ============================================================
# 1. EXPLICACIÓN CONCEPTUAL DEL MODELO
# ============================================================

print("============================================================")
print("MODELO OCULTO DE MARKOV - HMM")
print("============================================================")
print("""
Un Modelo Oculto de Markov, o HMM, permite representar procesos donde
existe una secuencia de estados reales que no se observa directamente.

En este caso, el estado oculto es el clima real:
- Soleado
- Nublado
- Lluvioso

La observación visible es el nivel de humedad:
- Humedad Baja
- Humedad Media
- Humedad Alta

El clima real no siempre se observa directamente, pero puede inferirse
a partir de señales observables como la humedad.

El modelo usa tres tipos de probabilidades:

1. Probabilidades iniciales:
   Indican la probabilidad de iniciar en cada estado climático.

2. Matriz de transición:
   Indica la probabilidad de pasar de un estado climático a otro.

3. Matriz de emisión:
   Indica la probabilidad de observar cierto nivel de humedad dado
   un estado climático real.
""")

# ============================================================
# 2. FORMULACIÓN DEL PROBLEMA
# ============================================================

print("============================================================")
print("FORMULACIÓN DEL PROBLEMA")
print("============================================================")
print("""
Problema:
Se desea simular el comportamiento del clima durante un periodo de 30 días.
El clima real no se observa directamente, por eso se considera un estado oculto.
En cambio, sí se puede observar el nivel de humedad registrado cada día.

Objetivo:
Construir un Modelo Oculto de Markov que permite simular la evolución diaria
del clima y generar observaciones de humedad asociadas a cada estado.

Pertinencia del HMM:
El problema es adecuado para un HMM porque el clima real funciona como una
variable oculta, mientras que la humedad funciona como una señal observable
que depende probabilísticamente del estado climático.
""")

# ============================================================
# 3. DEFINICIÓN DEL MODELO HMM
# ============================================================

class ClimaHMM:
    """
    Modelo Oculto de Markov para simular el clima real a partir de
    observaciones de humedad.
    """

    def __init__(self):
        # Estados ocultos: clima real
        self.estados = ["Soleado", "Nublado", "Lluvioso"]

        # Observaciones visibles: humedad registrada
        self.observaciones = ["Humedad Baja", "Humedad Media", "Humedad Alta"]

        # Probabilidad inicial de cada estado climático
        self.prob_inicial = np.array([0.6, 0.3, 0.1])

        # Matriz de transición
        # Filas: estado actual
        # Columnas: estado siguiente
        self.matriz_transicion = np.array([
            [0.7, 0.2, 0.1],  # Soleado -> Soleado, Nublado, Lluvioso
            [0.3, 0.4, 0.3],  # Nublado -> Soleado, Nublado, Lluvioso
            [0.2, 0.3, 0.5]   # Lluvioso -> Soleado, Nublado, Lluvioso
        ])

        # Matriz de emisión
        # Filas: estado climático real
        # Columnas: observación de humedad
        self.matriz_emision = np.array([
            [0.7, 0.2, 0.1],  # Si está Soleado
            [0.2, 0.5, 0.3],  # Si está Nublado
            [0.1, 0.3, 0.6]   # Si está Lluvioso
        ])

        # Colores para gráficos
        self.colores_estados = {
            "Soleado": "#FFD700",
            "Nublado": "#87CEFA",
            "Lluvioso": "#00008B"
        }

        self.colores_observaciones = {
            "Humedad Baja": "#90EE90",
            "Humedad Media": "#3CB371",
            "Humedad Alta": "#006400"
        }

    def validar_matrices(self):
        """
        Verifica que las probabilidades iniciales, matriz de transición
        y matriz de emisión estén correctamente normalizadas.
        """

        print("============================================================")
        print("VALIDACIÓN DE PROBABILIDADES")
        print("============================================================")

        print("Suma de probabilidades iniciales:", self.prob_inicial.sum())
        print("Suma por fila de la matriz de transición:", self.matriz_transicion.sum(axis=1))
        print("Suma por fila de la matriz de emisión:", self.matriz_emision.sum(axis=1))

        if np.isclose(self.prob_inicial.sum(), 1) and \
           np.allclose(self.matriz_transicion.sum(axis=1), 1) and \
           np.allclose(self.matriz_emision.sum(axis=1), 1):
             print("\nTodas las probabilidades son válidas. Cada distribución suma 1.")
        else:
             print("\nHay un error: alguna distribución de probabilidad no suma 1.")

    def mostrar_matrices(self):
        """
        Muestra las matrices del modelo en formato tabular.
        """

        print("============================================================")
        print("PROBABILIDADES INICIALES")
        print("============================================================")

        df_inicial = pd.DataFrame({
            "Estado": self.estados,
            "Probabilidad inicial": self.prob_inicial
        })

        display(df_inicial)

        print("============================================================")
        print("MATRIZ DE TRANSICIÓN")
        print("============================================================")

        df_transicion = pd.DataFrame(
            self.matriz_transicion,
            index=[f"Desde {estado}" for estado in self.estados],
            columns=[f"Hacia {estado}" for estado in self.estados]
        )

        display(df_transicion)

        print("============================================================")
        print("MATRIZ DE EMISIÓN")
        print("============================================================")

        df_emision = pd.DataFrame(
            self.matriz_emision,
            index=[f"Estado {estado}" for estado in self.estados],
            columns=self.observaciones
        )

        display(df_emision)

    def simular(self, dias=30):
        """
        Genera una simulación de estados ocultos y observaciones durante n días.
        """

        secuencia_estados = []
        secuencia_observaciones = []

        # Selección del estado inicial
        estado_actual_idx = np.random.choice(
            len(self.estados),
            p=self.prob_inicial
        )

        # Generación de la primera observación
        observacion_actual_idx = np.random.choice(
            len(self.observaciones),
            p=self.matriz_emision[estado_actual_idx]
        )

        secuencia_estados.append(self.estados[estado_actual_idx])
        secuencia_observaciones.append(self.observaciones[observacion_actual_idx])

        # Generación de los días siguientes
        for _ in range(1, dias):
            estado_actual_idx = np.random.choice(
                len(self.estados),
                p=self.matriz_transicion[estado_actual_idx]
            )

            observacion_actual_idx = np.random.choice(
                len(self.observaciones),
                p=self.matriz_emision[estado_actual_idx]
            )

            secuencia_estados.append(self.estados[estado_actual_idx])
            secuencia_observaciones.append(self.observaciones[observacion_actual_idx])

        df_resultado = pd.DataFrame({
            "Día": np.arange(1, dias + 1),
            "Estado Oculto": secuencia_estados,
            "Observación": secuencia_observaciones
        })

        return df_resultado

    def calcular_distribucion_estacionaria(self):
        """
        Calcula la distribución estacionaria teórica de la cadena de Markov.
        Esta distribución indica el comportamiento esperado a largo plazo.
        """

        valores, vectores = np.linalg.eig(self.matriz_transicion.T)

        indice = np.argmin(np.abs(valores - 1))
        estacionaria = np.real(vectores[:, indice])
        estacionaria = estacionaria / estacionaria.sum()

        return estacionaria

    def calcular_probabilidades_observaciones_teoricas(self):
        """
        Calcula la probabilidad teórica de cada observación a largo plazo.
        """

        estacionaria = self.calcular_distribucion_estacionaria()
        prob_obs = estacionaria @ self.matriz_emision

        return prob_obs

# ============================================================
# 4. CREACIÓN Y VALIDACIÓN DEL MODELO
# ============================================================

hmm = ClimaHMM()

hmm.validar_matrices()
hmm.mostrar_matrices()

# ============================================================
# 5. SIMULACIÓN PRINCIPAL DE 30 DÍAS
# ============================================================

dias_simulacion = 30
df_simulacion = hmm.simular(dias=dias_simulacion)

print("============================================================")
print("SIMULACIÓN PRINCIPAL DE 30 DÍAS")
print("============================================================")

display(df_simulacion)

print("Primeras 10 observaciones:")
display(df_simulacion.head(10))

# ============================================================
# 6. ESTADÍSTICAS DE LA SIMULACIÓN DE 30 DÍAS
# ============================================================

print("============================================================")
print("ESTADÍSTICAS DE LA SIMULACIÓN DE 30 DÍAS")
print("============================================================")

conteo_estados = df_simulacion["Estado Oculto"].value_counts().reindex(hmm.estados, fill_value=0)
porcentaje_estados = df_simulacion["Estado Oculto"].value_counts(normalize=True).reindex(hmm.estados, fill_value=0) * 100

df_estados_30 = pd.DataFrame({
    "Estado Oculto": hmm.estados,
    "Cantidad de días": conteo_estados.values,
    "Porcentaje observado": porcentaje_estados.values
})

display(df_estados_30)

conteo_observaciones = df_simulacion["Observación"].value_counts().reindex(hmm.observaciones, fill_value=0)
porcentaje_observaciones = df_simulacion["Observación"].value_counts(normalize=True).reindex(hmm.observaciones, fill_value=0) * 100

df_observaciones_30 = pd.DataFrame({
    "Observación": hmm.observaciones,
    "Cantidad de días": conteo_observaciones.values,
    "Porcentaje observado": porcentaje_observaciones.values
})

display(df_observaciones_30)

# ============================================================
# 7. MATRIZ DE CONFUSIÓN ESTADO VS OBSERVACIÓN
# ============================================================

print("============================================================")
print("MATRIZ DE CONFUSIÓN: ESTADO OCULTO VS OBSERVACIÓN")
print("============================================================")

matriz_confusion = pd.crosstab(
    df_simulacion["Estado Oculto"],
    df_simulacion["Observación"]
)

matriz_confusion = matriz_confusion.reindex(
    index=hmm.estados,
    columns=hmm.observaciones,
    fill_value=0
)

display(matriz_confusion)

# ============================================================
# 8. GRÁFICOS DE LA SIMULACIÓN DE 30 DÍAS (VERSIÓN MEJORADA)
# ============================================================

print("============================================================")
print("GRÁFICOS DE LA SIMULACIÓN DE 30 DÍAS")
print("============================================================")

dias_x = df_simulacion["Día"]

# Gráficos apilados que comparten el eje X para facilitar la comparación
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 5), sharex=True)
fig.subplots_adjust(hspace=0.1)

# 1. Gráfico superior: Estados Ocultos
colores_estados = [hmm.colores_estados[e] for e in df_simulacion["Estado Oculto"]]
ax1.bar(dias_x, [1] * dias_simulacion, color=colores_estados, edgecolor="black")
ax1.set_title("Evolución diaria: Clima Real (Arriba) vs Humedad Registrada (Abajo)")
ax1.set_yticks([])
handles_estados = [
    plt.Rectangle((0, 0), 1, 1, color=hmm.colores_estados[e])
    for e in hmm.estados
]
ax1.legend(handles_estados, hmm.estados, loc="upper right")

# 2. Gráfico inferior: Observaciones
colores_obs = [hmm.colores_observaciones[o] for o in df_simulacion["Observación"]]
ax2.bar(dias_x, [1] * dias_simulacion, color=colores_obs, edgecolor="black")
ax2.set_xlabel("Día")
ax2.set_yticks([])
ax2.set_xticks(dias_x)
handles_obs = [
    plt.Rectangle((0, 0), 1, 1, color=hmm.colores_observaciones[o])
    for o in hmm.observaciones
]
ax2.legend(handles_obs, hmm.observaciones, loc="upper right")

plt.show()

# 3. Matriz de confusión
plt.figure(figsize=(8, 5))
sns.heatmap(matriz_confusion, annot=True, cmap="Blues", fmt="d")
plt.title("Matriz de Confusión: Estado Oculto vs Observación")
plt.ylabel("Estado Oculto")
plt.xlabel("Observación")
plt.show()

# ============================================================
# 9. CÁLCULO DE PROBABILIDADES TEÓRICAS
# ============================================================

print("============================================================")
print("PROBABILIDADES TEÓRICAS DE LARGO PLAZO")
print("============================================================")

prob_estados_teoricas = hmm.calcular_distribucion_estacionaria()
prob_observaciones_teoricas = hmm.calcular_probabilidades_observaciones_teoricas()

df_prob_estados_teoricas = pd.DataFrame({
    "Estado Oculto": hmm.estados,
    "Probabilidad teórica": prob_estados_teoricas,
    "Porcentaje teórico": prob_estados_teoricas * 100,
    "Días esperados en 30 días": prob_estados_teoricas * 30
})

display(df_prob_estados_teoricas)

df_prob_observaciones_teoricas = pd.DataFrame({
    "Observación": hmm.observaciones,
    "Probabilidad teórica": prob_observaciones_teoricas,
    "Porcentaje teórico": prob_observaciones_teoricas * 100,
    "Días esperados en 30 días": prob_observaciones_teoricas * 30
})

display(df_prob_observaciones_teoricas)

# ============================================================
# 10. MUCHAS SIMULACIONES DE 30 DÍAS PARA ANALIZAR CONVERGENCIA
# ============================================================

print("============================================================")
print("ANÁLISIS DE CONVERGENCIA CON MUCHAS SIMULACIONES DE 30 DÍAS")
print("============================================================")

num_simulaciones = 10000
dias_por_simulacion = 30
total_dias_simulados = num_simulaciones * dias_por_simulacion

conteo_total_estados = {estado: 0 for estado in hmm.estados}
conteo_total_observaciones = {obs: 0 for obs in hmm.observaciones}

# Guardaremos la evolución de las probabilidades empíricas
historial_convergencia = []

for simulacion in range(1, num_simulaciones + 1):
    df_temp = hmm.simular(dias=dias_por_simulacion)

    for estado in df_temp["Estado Oculto"]:
        conteo_total_estados[estado] += 1

    for obs in df_temp["Observación"]:
        conteo_total_observaciones[obs] += 1

    # Guardar puntos de control para graficar convergencia
    if simulacion % 100 == 0:
        dias_acumulados = simulacion * dias_por_simulacion

        fila = {
            "Simulación": simulacion,
            "Días acumulados": dias_acumulados
        }

        for estado in hmm.estados:
            fila[f"Empírico {estado}"] = conteo_total_estados[estado] / dias_acumulados

        for obs in hmm.observaciones:
            fila[f"Empírico {obs}"] = conteo_total_observaciones[obs] / dias_acumulados

        historial_convergencia.append(fila)

df_convergencia = pd.DataFrame(historial_convergencia)

print(f"Número de simulaciones realizadas: {num_simulaciones}")
print(f"Días por simulación: {dias_por_simulacion}")
print(f"Total de días simulados: {total_dias_simulados}")

# ============================================================
# 11. COMPARACIÓN ENTRE RESULTADOS EMPÍRICOS Y TEÓRICOS
# ============================================================

print("============================================================")
print("COMPARACIÓN: ESTADOS OCULTOS")
print("============================================================")

prob_estados_empiricas = np.array([
    conteo_total_estados[estado] / total_dias_simulados
    for estado in hmm.estados
])

df_comparacion_estados = pd.DataFrame({
    "Estado Oculto": hmm.estados,
    "Probabilidad teórica": prob_estados_teoricas,
    "Probabilidad empírica": prob_estados_empiricas,
    "Diferencia absoluta": np.abs(prob_estados_teoricas - prob_estados_empiricas)
})

display(df_comparacion_estados)

print("============================================================")
print("COMPARACIÓN: OBSERVACIONES")
print("============================================================")

prob_observaciones_empiricas = np.array([
    conteo_total_observaciones[obs] / total_dias_simulados
    for obs in hmm.observaciones
])

df_comparacion_observaciones = pd.DataFrame({
    "Observación": hmm.observaciones,
    "Probabilidad teórica": prob_observaciones_teoricas,
    "Probabilidad empírica": prob_observaciones_empiricas,
    "Diferencia absoluta": np.abs(prob_observaciones_teoricas - prob_observaciones_empiricas)
})

display(df_comparacion_observaciones)

# ============================================================
# 12. GRÁFICOS DE CONVERGENCIA
# ============================================================

print("============================================================")
print("GRÁFICOS DE CONVERGENCIA")
print("============================================================")

plt.figure(figsize=(12, 6))

for i, estado in enumerate(hmm.estados):
    plt.plot(
        df_convergencia["Días acumulados"],
        df_convergencia[f"Empírico {estado}"],
        label=f"Empírico {estado}"
    )

    plt.axhline(
        y=prob_estados_teoricas[i],
        linestyle="--",
        label=f"Teórico {estado}"
    )

plt.title("Convergencia de probabilidades de estados ocultos")
plt.xlabel("Días acumulados simulados")
plt.ylabel("Probabilidad")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))

for i, obs in enumerate(hmm.observaciones):
    plt.plot(
        df_convergencia["Días acumulados"],
        df_convergencia[f"Empírico {obs}"],
        label=f"Empírico {obs}"
    )

    plt.axhline(
        y=prob_observaciones_teoricas[i],
        linestyle="--",
        label=f"Teórico {obs}"
    )

plt.title("Convergencia de probabilidades de observaciones")
plt.xlabel("Días acumulados simulados")
plt.ylabel("Probabilidad")
plt.legend()
plt.grid(True)
plt.show()

# ============================================================
# 13. GRÁFICOS COMPARATIVOS FINALES
# ============================================================

print("============================================================")
print("GRÁFICOS COMPARATIVOS FINALES")
print("============================================================")

df_grafico_estados = df_comparacion_estados.melt(
    id_vars="Estado Oculto",
    value_vars=["Probabilidad teórica", "Probabilidad empírica"],
    var_name="Tipo",
    value_name="Probabilidad"
)

plt.figure(figsize=(8, 5))
sns.barplot(
    data=df_grafico_estados,
    x="Estado Oculto",
    y="Probabilidad",
    hue="Tipo"
)
plt.title("Comparación entre probabilidades teóricas y empíricas de estados")
plt.ylim(0, 1)
plt.show()

df_grafico_observaciones = df_comparacion_observaciones.melt(
    id_vars="Observación",
    value_vars=["Probabilidad teórica", "Probabilidad empírica"],
    var_name="Tipo",
    value_name="Probabilidad"
)

plt.figure(figsize=(8, 5))
sns.barplot(
    data=df_grafico_observaciones,
    x="Observación",
    y="Probabilidad",
    hue="Tipo"
)
plt.title("Comparación entre probabilidades teóricas y empíricas de observaciones")
plt.ylim(0, 1)
plt.show()

# ============================================================
# 14. CONCLUSIONES PARA EL INFORME O VIDEO
# ============================================================

print("============================================================")
print("CONCLUSIONES")
print("============================================================")

print(f"""
1. Se construyó un Modelo Oculto de Markov para simular el clima real
   a partir de observaciones de humedad.

2. Los estados ocultos del modelo fueron:
   {hmm.estados}

3. Las observaciones visibles fueron:
   {hmm.observaciones}

4. Se realizó una simulación principal de {dias_simulacion} días, en la cual
   se generó una secuencia diaria de estados climáticos y observaciones
   de humedad.

5. Además, se realizaron {num_simulaciones} simulaciones independientes
   de {dias_por_simulacion} días cada una.

6. En total, se simularon {total_dias_simulados} días para analizar la
   convergencia de las probabilidades empíricas hacia los valores teóricos.

7. La distribución estacionaria teórica de los estados ocultos fue:
""")

for estado, prob in zip(hmm.estados, prob_estados_teoricas):
    print(f"   - {estado}: {prob * 100:.2f}%")

print("\n8. Las probabilidades empíricas obtenidas después de muchas simulaciones fueron:")

for estado, prob in zip(hmm.estados, prob_estados_empiricas):
    print(f"   - {estado}: {prob * 100:.2f}%")

print("""
9. Las diferencias entre los valores teóricos y empíricos son pequeñas,
   lo cual evidencia que al aumentar el número de simulaciones, los resultados
   tienden a aproximarse a las probabilidades teóricas del modelo.

10. Por lo tanto, el HMM construido permite representar adecuadamente un proceso
    donde el estado real no se observa directamente, pero puede analizarse mediante
    observaciones visibles relacionadas probabilísticamente.
""")
```
